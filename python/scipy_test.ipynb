```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# Generate sine wave data - pure, noisy and filtered
x = np.linspace(0, 2 * np.pi, 100)
y_pure = np.sin(x)
y_noisy = np.sin(x) + np.random.random(100) * 0.4 - 0.2
y_savgol = savgol_filter(y_noisy, 51, 3)

# Create new figure and set axis limits
plt.figure(figsize=(16, 9))
plt.xlim(0, 2 * np.pi)
plt.ylim(-1.2, 1.2)

# Plot data
plt.plot(x, y_pure, label = 'pure', color='green', linewidth = 1, linestyle='dashed')
plt.plot(x, y_noisy, label = 'noisy', color='blue', linestyle='none', marker='o', markerfacecolor='blue', markersize=5)
plt.plot(x, y_savgol, label = 'savgol', color='red', linewidth = 2)

# Add title, axis labels and legend
plt.title('Savgol Filter Test')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

# Show figure
plt.show()
```
